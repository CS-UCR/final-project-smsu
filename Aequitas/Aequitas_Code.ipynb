```python
import aequitas
import pandas as pd
```

The following is the Aequitas Output for the neural network model from the research article.

```python
df_nn = pd.read_csv('neural_network_copy.csv')
```

```python
from aequitas.preprocessing import preprocess_input_df
    
    # double-check that categorical columns are of type 'string'
df_nn['gender'] = df_nn['gender'].astype(str)
df_nn['Residence_type'] = df_nn['Residence_type'].astype(str)
    
df, _ = preprocess_input_df(df_nn)
```

```python
from aequitas.group import Group
g = Group()
xtab, _ = g.get_crosstabs(df)
```

```python
from aequitas.plotting import Plot
aqp = Plot()
fpr_plot = aqp.plot_group_metric(xtab, 'fpr')
```

```python
from aequitas.bias import Bias
b = Bias()
bdf = b.get_disparity_major_group(xtab,
                               original_df=df,
                               alpha=0.05,
                               check_significance=False)
```

```python
fpr_disparity = aqp.plot_disparity(bdf, group_metric='fpr_disparity', 
                                       attribute_name='gender')
```

```python
fpr_disparity = aqp.plot_disparity(bdf, group_metric='fpr_disparity', 
                                       attribute_name='age')
```

```python
fpr_disparity = aqp.plot_disparity(bdf, group_metric='fpr_disparity', 
                                       attribute_name='Residence_type')
```

```python
from aequitas.fairness import Fairness
f = Fairness()
fdf = f.get_group_value_fairness(bdf)
```

```python
fpr_fairness = aqp.plot_fairness_group(fdf, group_metric='fpr', title=True)
```

```python
fnr_disparity = aqp.plot_disparity(bdf, group_metric='fnr_disparity', 
                                       attribute_name='gender')
```

```python
fnr_disparity = aqp.plot_disparity(bdf, group_metric='fnr_disparity', 
                                       attribute_name='age')
```

```python
fnr_disparity = aqp.plot_disparity(bdf, group_metric='fnr_disparity', 
                                       attribute_name='Residence_type')
```
